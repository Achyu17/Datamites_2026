a)basic random operations
b)works in list,numbers and sequence 
c)Slower for large arrays not vectorized

#Random package numpy 
a)Optimized for array,matrices,data science
b)Works in vectorized operations for big data 
c)Offers advance distribution (normal,poission,binomial)
