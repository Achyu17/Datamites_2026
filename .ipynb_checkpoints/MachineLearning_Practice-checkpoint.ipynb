```python
#Numpy Indexing and Selection
mat = np.arange(1,26).reshape(5,5)
mat
```

```python
print(mat[3, 4])
```

```python
result = np.array([2, 7, 12]).reshape(-1, 1)
print(result)
```

```python
result = mat[4, :]
result
```

```python
total_sum = mat.sum()
print(total_sum)
```

```python
std_val=mat.std()
print(std_val)
```

```python
#Concatenate 2 matrix using Numpy
import numpy as np

A = np.array([[13, 21],
              [34, 42]])

B = np.array([[25, 6],
              [7, 18]])

#print(np.concatenate((A,B),axis=0))
print(np.concatenate((A,B),axis=1))
```

```python
#Sin,cos,log using Numpy
arr = np.array([0, np.pi/6, np.pi/4, np.pi/2, np.pi])
print("sin:", np.sin(arr))
print("cos:", np.sin(arr))
print("log10:", np.log10(arr))
print("log2:", np.log2(arr))
```

```python
#dot product using Numpy
a=np.array([7,8,11])
b=np.array([17,3,1])
res=np.dot(a,b)
print(res)
```

```python
#linear Algebra
A = np.array([[7, 1],
              [3, 4]])

det = np.linalg.det(A)
det
```

```python
#Calculus
import sympy as sp

x = sp.symbols('x')
f = x**3 + 4*x + 7

# Derivative
df = sp.diff(f, x)
print("f(x) =", f)
print("f'(x) =", df)
```
